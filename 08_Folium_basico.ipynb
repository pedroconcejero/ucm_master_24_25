```python
import folium 
```

```python
m = folium.Map(location = coordenadas_ucm_docu,
    zoom_start=16)

m
```

```python
triollo = [42.92540955744316, -4.681189734719546]
```

```python
m = folium.Map(location = triollo,
               zoom_start=14
              )

m
```

```python
folium.TileLayer('Carto DB Positron').add_to(m)
```

```python
m
```

```python
folium.TileLayer('Stamen Terrain', 
                 name = 'Stamen Toner', 
                 attr = 'Map tiles by Stamen Design, under CC BY 3.0. Data by OpenStreetMap, under ODbL.').add_to(m)
```

```python
m
```

```python
m = folium.Map(location = coordenadas_ucm_docu,
    zoom_start=14, 
    tiles='Carto DB Positron' 
) 
tooltip = 'UCM docu'
folium.Marker(coordenadas_ucm_docu, 
              popup = 'UCM', 
              tooltip = tooltip).add_to(m) 
```

```python
m
```

```python
m = folium.Map(location = coordenadas_ucm_docu,
    zoom_start=14 
) 
tooltip = 'UCM docu'
folium.Marker(coordenadas_ucm_docu, 
              popup = 'UCM', 
              tooltip = tooltip).add_to(m) 
```

```python
m
```
