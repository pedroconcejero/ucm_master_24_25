```python
m = folium.Map(location = coordenadas_ucm_docu,
    zoom_start=14, 
    tiles='Carto DB Positron' 
) 
tooltip = 'UCM docu'
folium.Marker(coordenadas_ucm_docu, 
              popup = 'Máster DS', 
              tooltip = tooltip).add_to(m) 
```

```python
m
```

```python
map = folium.Map(location = coordenadas_ucm_docu,
    zoom_start=14 
) 
```

```python
map
```

```python
tooltip = "Haz click para ver la dirección!"

folium.Marker(
    [40.439889422283926, -3.711783660232115], 
    popup="<a href='https://www.ucm.es/estudios/grado-turismo' target = '_blank'>UCM Turismo</a>", 
    tooltip=tooltip,
     icon=folium.Icon(color="red", icon="info-sign"),
).add_to(map)
folium.Marker(
    [40.43847105761711, -3.7014854890675513], 
    popup="<a href='https://documentacion.ucm.es/' target = '_blank'>UCM Documentación</a>", 
    tooltip=tooltip,
    icon=folium.Icon(color="blue"),
).add_to(map)

map
```
